```python
from azure.cognitiveservices.vision.customvision.prediction import CustomVisionPredictionClient
import os
def tag_image(img_dir):
    prediction_key = "8af2c213f6214d2ba569b36069be92f7"
    # Now there is a trained endpoint that can be used to make a prediction

    predictor = CustomVisionPredictionClient(prediction_key, endpoint=ENDPOINT)

    # Open the sample image and get back the prediction results.
    od_output=[]

    with open(img_dir, mode="rb") as test_data:
        results = predictor.predict_image('3796f0f6-b4e6-49e5-8ee5-f818aeee18cf', test_data)

    # Display the results.
        for prediction in results.predictions:
            if prediction.probability > 0.7:
                tag, prob, left, top, width, height = prediction.tag_name,prediction.probability,prediction.bounding_box.left, prediction.bounding_box.top, prediction.bounding_box.width, prediction.bounding_box.height
                od_output.append([tag, prob, left, top, width, height])
            
    return od_output
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=['image_dir','tag','prob','left','top','width','height'])
```

```python
new_df=df.copy()
```

```python
df['image_dir'] = os.listdir('safety_images')
```

```python
from tqdm import tqdm_notebook as tqdm
```

```python
for ind,row in tqdm(df.iterrows(),total=df.shape[0]):
    output_list=tag_image(os.path.join('safety_images',row['image_dir']))
    if len(output_list)>1:
        for one_list in output_list:
            tag, prob, left, top, width, height = one_list
            new_df=new_df.append(pd.DataFrame({'image_dir': row['image_dir'],
                           'tag': [tag],
                           'prob': [prob],
                           'left': [left],
                           'top': [top],
                           'width': [width],
                           'height': [height]}))
            
    elif len(output_list)==1:
        row['tag'], row['prob'], row['left'], row['top'], row['width'], row['height'] = tag, prob, left, top, width, height
    else:
        print('Error')
        
        
```
